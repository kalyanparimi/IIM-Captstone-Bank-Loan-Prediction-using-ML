> ### Create a function for easy report printing

```python
def print_classification_report(y_true, y_prediction, *type_of_report):
    accuracy = accuracy_score(y_true, y_prediction)
    precision = precision_score(y_true, y_prediction)
    recall = recall_score(y_true, y_prediction)
    f1 = f1_score(y_true, y_prediction)
    
    if type_of_report:
        print(color.BOLD + 'Classification Report on %s:' %type_of_report + color.END)
    else:
        print('Classification Report:')
    print()
    print('---------------------------------------')
    print('Accuracy: %s' %accuracy)
    print('Precision: %s' %precision)
    print(color.BOLD + 'Recall: %s' %recall + color.END)
    print('F1 score: %s' %f1)
    print('---------------------------------------')

# A class for pretty printing
class color:
    PURPLE = '\033[95m'
    CYAN = '\033[96m'
    DARKCYAN = '\033[36m'
    BLUE = '\033[94m'
    GREEN = '\033[92m'
    YELLOW = '\033[93m'
    RED = '\033[91m'
    BOLD = '\033[1m'
    UNDERLINE = '\033[4m'
    END = '\033[0m'
```

**Create the first set of training and test data on imbalanced data**

```python
df = pd.get_dummies(data, columns=['Gender','Occupation'], drop_first = True)
df.head()
```

>**Creating a model with unbalanced data and measuring metrics**

```python
X_original = df.iloc[:,1:]
y_original = df.iloc[:,0]
```

```python
X_train_orig, X_test_orig, y_train_orig, y_test_orig = train_test_split(X_original,y_original, shuffle = ['True'])
```

```python
X_train_orig, X_test_orig, y_train_orig, y_test_orig = train_test_split(X_original,y_original, shuffle = ['True'], stratify=y_original)
```

```python
clf = DecisionTreeClassifier(max_depth = 5)
clf.fit(X_train_orig, y_train_orig)
y_prediction_orig = clf.predict(X_test_orig)
```

```python
print_classification_report(y_test_orig, y_prediction_orig, 'Original data')
```

```python
X_under_train, y_under_train = NearMiss().fit_resample(X_under_train, y_under_train)
```
